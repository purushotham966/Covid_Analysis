```python
def nice_axes(ax):
    ax.set_facecolor('.8')
    ax.tick_params(labelsize=7, length=0)
    ax.grid(True, axis='x', color='white')
    ax.set_axisbelow(True)
    [spine.set_visible(False) for spine in ax.spines.values()]
```

```python
def prepare(d,a):
    d=d.reset_index()
    d.index=d.index*a
    i=d.index[-1]+1
    df_h=d.reindex(range(i))
    df_h["index"]=df_h["index"].fillna(method="ffill")
    df_h=df_h.set_index("index")
    df_rank_expanded = df_h.rank(axis=1, method='first')
    #print(df_h)
    df_h=df_h.interpolate()
    df_rank_expanded = df_rank_expanded.interpolate()
    return(df_h,df_rank_expanded)
df1,df_rank_expanded=prepare(df,3)
df_rank_expanded
```

```python
df1.iloc[1]
```

```python
colors = plt.cm.Dark2(range(6))
df1.iloc[1]
```

```python
labels = df1.columns
```

```python
from matplotlib.animation import FuncAnimation

def init():
    ax.clear()
    nice_axes(ax)
    ax.set_ylim(.2, 6.8)

def update(i):
    for bar in ax.containers:
        bar.remove()
    y = df_rank_expanded.iloc[i]
    width = df1.iloc[i]
    #print(df1)
    ax.barh(y=y, width=width, color=colors, tick_label=labels)
    date_str = df1.index[i]
    ax.set_title(f'COVID-19 confirmed by Country - {date_str}', fontsize='smaller')
    
fig = plt.Figure(figsize=(5, 3), dpi=144)
ax = fig.add_subplot()
anim = FuncAnimation(fig=fig, func=update, init_func=init, frames=len(df1), 
                     interval=100, repeat=False)
```

```python
from IPython.display import HTML
html = anim.to_html5_video()
HTML(html)
```
